**Diabetes Dataset EDA:**


**Univariate Analysis:**

Analyze individual variables (e.g., histograms, box plots).

**Bivariate Analysis:**

Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).

**Multivariate Analysis:**

Investigate interactions among multiple variables (e.g., pair plots, 3D plots).

```python
import pandas as pd
file_path = r'C:\DSA ICTKERALA\Intermediate Assesment\diabetes_model.csv'
df = pd.read_csv(file_path)
print(df.head())
```

**Univariate Analysis**


```python
# Histograms
import matplotlib.pyplot as plt
df.hist(figsize=(12, 10), bins=15)
plt.suptitle("Histograms of Variables", fontsize=16)
plt.show()
```

```python
# Box Plots
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.boxplot(data=df)
plt.title("Box Plots of Variables", fontsize=16)
plt.xticks(rotation=45)
plt.show()

```

```python
# Summary Statistics
print("Summary Statistics:\n", df.describe())
```

**BIVARIATE ANALYSIS**

```python
# Scatter Plot: Glucose vs BMI
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Glucose', y='BMI', data=df)
plt.title("Scatter Plot of Glucose vs BMI", fontsize=14)
plt.show()
```

```python

# Correlation Matrix and Heatmap
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.show()
```

```python

# Box Plot: Outcome vs Glucose
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title("Box Plot of Outcome vs Glucose", fontsize=14)
plt.show()
```

**MULTIVARIATE ANALYSIS**

```python
# Pair Plot
import seaborn as sns
sns.pairplot(df, hue='Outcome', diag_kind='hist', corner=True)
plt.suptitle("Pair Plot of Variables", fontsize=16, y=1.02)
plt.show()
```

```python
# 3D Plot: Age, Glucose, and BMI
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Glucose'], df['BMI'], c=df['Outcome'], cmap='viridis')
ax.set_xlabel('Age')
ax.set_ylabel('Glucose')
ax.set_zlabel('BMI')
plt.title("3D Plot of Age, Glucose, and BMI", fontsize=16)
plt.show()
```
